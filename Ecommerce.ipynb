```python
#Sales Trend over time

#Ensure date is in datetime format
sales_df['date']=pd.to_datetime(sales_df['date'])

daily_sales=sales_df.groupby('date').agg({'quantity':'sum'}).reset_index()

plt.figure(figsize=(10,6))
plt.plot(daily_sales['date'],daily_sales['quantity'],marker='o',linestyle='-',color='b')
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Quantity')
plt.xticks(rotation=160)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#SALES DISTRIBUTION BY PRODUCT CATEGORY
#Merge sales data with product data to get category info
sales_with_category=sales_df.merge(products_df[['product_id','category']],on='product_id',how='left')


#Aggregate sales by category
category_sales = sales_with_category.groupby('category').agg({'quantity':'sum'}).reset_index()

# Plot sales dist by product category

plt.figure(figsize=(10,6))
sns.barplot(x='category',y='quantity',data=category_sales,palette='viridis',hue='category',legend=False)
plt.title('Sales Distribution by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
# REGIONAL SALES PERFORMANCE

#Merge sales df with products df to get the category

sales_with_category = sales_df.merge(products_df[['product_id','category']], on='product_id',how='left')

# Grouping data by region and category

regional_sales_summary = sales_with_category.groupby(['region','category']).agg({'quantity':'sum'}).reset_index()

# Create the pivot table for heatmap
pivot_table = regional_sales_summary.pivot(index='region', columns='category', values='quantity')

#Plot the heatmap

plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True,fmt='g',cmap='viridis')
plt.title(' Sales Quantity by Region and Category')
plt.show()
```

```python
# CORRELATION ANALYSIS BTN QTY AND PRICES

# Merge sales with product prices
sales_with_price = sales_df.merge(products_df[['product_id','price']], on='product_id',how='left')


# Calculate correlarion matrix

correlation_matrix = sales_with_price[['quantity','price']].corr()

#Plot the correlation matrix

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
# CORRELATION BTN SALES AND CUSTOMER AGES

# Merge sales df with customers df

sales_with_age=sales_df.merge(customers_df[['customer_id','age']],on='customer_id',how='left')

# Calculate correlation matrix
correlation_matrix = sales_with_age[['quantity','age']].corr()

# Plot correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',center=0)
plt.title('Correlation between Quantity and Age')
plt.show()
```

```python
# RELATIONSHIP BTN AGE AND CATEGORIES: BOX PLOT

sales_category = sales_df.merge(products_df[['product_id','category']],on='product_id',how='left')
category_ages=sales_category.merge(customers_df[['customer_id','age']],on='customer_id',how='left')

plt.figure(figsize=(10,6))
sns.boxplot(x='category',y='age',data=category_ages)
plt.title('Age Distribution By Product Category')
plt.xticks(rotation=45) #Rotate category labels
plt.show()
```

```python
# RELATIONSHIP BTN AGE AND CATEGORIES: Violin plot

sales_category = sales_df.merge(products_df[['product_id','category']],on='product_id',how='left')
category_ages=sales_category.merge(customers_df[['customer_id','age']],on='customer_id',how='left')

plt.figure(figsize=(10,6))
sns.violinplot(x='category',y='age',data=category_ages)
plt.title('Age Distribution By Product Category')
plt.xticks(rotation=45) #Rotate category labels
plt.show()
```

```python
# DESCRIPTIVE RELATIONSHIP
sales_category = sales_df.merge(products_df[['product_id','category']],on='product_id',how='left')
category_ages=sales_category.merge(customers_df[['customer_id','age']],on='customer_id',how='left')

age_by_category= category_ages.groupby('category')['age'].describe()
print(age_by_category)

```

```python
# RELATIONSHIP BTN AGE AND CATEGORIES: ANOVA,,ANALYSIS OF VARIANCE
# high f value and low p value means that the difference in mean ages across categories is quite significant.
#Reverse applies


import scipy.stats as stats


sales_category = sales_df.merge(products_df[['product_id','category']],on='product_id',how='left')
category_ages=sales_category.merge(customers_df[['customer_id','age']],on='customer_id',how='left')

grouped= category_ages.groupby('category')['age'].apply(list)

#perform anova
f_val,p_val = stats.f_oneway(*grouped)
print(f'ANOVA F-Value:{f_val},p-value: {p_val}')
```

```python
# CUSTOMER LIFETIME VALUE

#Merge sales and product data to get price info

sales_with_price = sales_df.merge(products_df[['product_id','price']],on='product_id', how='left')

# Calculate total purchase value per customer

sales_with_price['total_purchase_value'] = sales_with_price['quantity']*sales_with_price['price']
customer_purchase_value = sales_with_price.groupby('customer_id')['total_purchase_value'].sum().reset_index()

#Calc number of transactions per customer
transaction_counts = sales_with_price.groupby('customer_id')['transaction_id'].count().reset_index()
transaction_counts.rename(columns={'transaction_id':'num_transactions'}, inplace=True)

#Average purchase value per customer

customer_purchase_value = customer_purchase_value.merge(transaction_counts, on='customer_id')
customer_purchase_value['avg_purchase_value']= customer_purchase_value['total_purchase_value']/customer_purchase_value['num_transactions']

#Estimate purchase frequency(assuming customer lifespan as the no. of months or years if available)
#For now we'll use the number of transwactions as the frequency

customer_purchase_value['purchase_frequency']= customer_purchase_value['num_transactions']

# If customer lifespan is available,, we would include it in the calc:: for now assume lifespan=1,eg 1 yeasr

customer_purchase_value['customer_lifespan']=1

# Calculate CLV
customer_purchase_value['CLV']= customer_purchase_value['avg_purchase_value']*customer_purchase_value['purchase_frequency']*customer_purchase_value['customer_lifespan']
print(customer_purchase_value.head())
```

```python
# Monthly sales Aggregate

#Convert the date column to datetime if it isnt already

sales_with_price['date'] = pd.to_datetime(sales_with_price['date'])

 #Extract month and year from date
sales_with_price['year_month'] = sales_with_price['date'].dt.to_period('M')

#Calc monthly sales
monthly_sales = sales_with_price.groupby('year_month')['total_purchase_value'].sum().reset_index()

print (sales_with_price.head())
```
